```python
def scrpping(prod_link):
    # install and import packages

#!pip install html5lib
    import pandas as pd, warnings, requests
    from urllib.request import urlopen
    from bs4 import BeautifulSoup
    import html5lib
    prod_soup=BeautifulSoup(requests.get(prod_link).content)
    prd_container=prod_soup.find_all('div', {'class':"_1YokD2 _3Mn1Gg col-8-12"})
    
    if len(prd_container[0].findAll('span', {'class': "B_NuCI"})[0].text.split())>0:
        prd_name=prd_container[0].findAll('span', {'class': "B_NuCI"})[0].text
    else:
        prd_name='Product name missing'
    
    
    cost=prd_container[0].findAll('div', {'class':"_30jeq3 _16Jk6d"})[0].text
    
    if len(prd_container[0].findAll('span', {'class':"_2_R_DZ"})[0].text.split())>0:
        
        rattings=prd_container[0].findAll('span', {'class':"_2_R_DZ"})[0].text.split()[0]
    else:
            rattings="missining"
    if len(prd_container[0].findAll('span', {'class':"_2_R_DZ"})[0].text.split())>2:
        reviews=prd_container[0].findAll('span', {'class':"_2_R_DZ"})[0].text.split()[3]
    else:
        reviews="missing"
        
    
    stars=prd_container[0].findAll('div', {'class': "_3LWZlK"})[0].text
        
    
    comment_header_list=[]
    for i in prd_container[0].findAll('p', {'class':"_2-N8zT"}):
        comment_header_list.append(i.text)
    cust_name=[]
        
    for cust in prd_container[0].findAll('p', {'class': "_2sc7ZR _2V5EHH"}):
        cust_name.append(cust.text)

        #comment

    comment_list=[ i.text for i in prd_container[0].find_all("div", {'class':"t-ZTKy"})]
    my_dict={}
    my_dict["Product_name"]=prd_name
    my_dict["cost"]=cost
    my_dict["rattings"]=rattings
    my_dict["reviews"]=reviews
    my_dict['stars']=stars
    my_dict["Customer_name"]=cust_name
    my_dict["Review_header"]=comment_header_list
    my_dict["Comment"]=comment_list
        
    return(pd.DataFrame(my_dict))
    
    

    
        

        

        

def Flip_kart_review(url,product):
    import pandas as pd, warnings, requests
    from urllib.request import urlopen
    from bs4 import BeautifulSoup
    import html5lib
    
    
    q=product
    url=f"https://www.flipkart.com/search?q="+q
    soup1=BeautifulSoup( urlopen(url).read(), 'html.parser')
    container1=soup1.find_all('div', {'class':"_13oc-S"})
    links=[]
    for i in container1:
        links.append('https://www.flipkart.com'+i.div.a['href'])
    #df=pd.DataFrame()
    df=pd.DataFrame()
    for i in links:
        df=pd.concat([df, scrpping(i)])

       
        
            
        
        
        
    return df
        


        
```

```python
realme=Flip_kart_review(f"https://www.flipkart.com/search?q=", 'realme')
apple=Flip_kart_review(f"https://www.flipkart.com/search?q=", 'applemobile')
poco=Flip_kart_review(f"https://www.flipkart.com/search?q=", "poco")
samsung=Flip_kart_review(f"https://www.flipkart.com/search?q=", 'samsung')
oppo=Flip_kart_review(f"https://www.flipkart.com/search?q=", 'oppo')
lava=Flip_kart_review(f"https://www.flipkart.com/search?q=", "lavaphone")
lenov=Flip_kart_review(f"https://www.flipkart.com/search?q=", "lenovophone")
nothing=Flip_kart_review(f"https://www.flipkart.com/search?q=", "nothingphone")
one_plus=Flip_kart_review(f"https://www.flipkart.com/search?q=", "one_plus")
```
